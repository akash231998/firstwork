```python
img1=cv2.imread('box.png')
img2=cv2.imread('box_in_scene.png')
```

```python
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img1,cv2.COLOR_BGR2RGB))
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(img2,cv2.COLOR_BGR2RGB))
```

```python
import time
sift=cv2.xfeatures2d.SIFT_create()
gray1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
gray2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)

start_time=time.time()
kp1,des1=sift.detectAndCompute(gray1,None)
kp2,des2=sift.detectAndCompute(gray2,None)
print("Elapsed Time:%.6fs"%(time.time()-start.time))
print("Image1=%d features detected" %des1.shape[0])
print("Image2=%d features detected" %des2.shape[0])
```

```python
bf=cv2.BFMatcher(cv2.NORM_L2)
matches=bf.atch(des1,des2,k=2)
print("%d matches" %len(matches))
good_matches=[]
for m,n in matches:
    if m.distance<0.75*n.distance:
        good_matches.append(m)
print('%d matches' %len(good_matches))
```

```python
img3=cv2.drawMatches(img1,kp1,img2,kp2,good_matches,None)
plt.imshow(cv2.cvtColor(img3,cv2.COLOR_BGR2RGB))
```
